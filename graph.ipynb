```python
import matplotlib.pyplot as plt

# Datos
algorithms = ["DFS", "BFS", "A*h1", "A*h2", "GSh1", "GSh2"]
nodes = [681, 127465, 80249, 59158, 33553, 27144]
time = [0.212, 31.195, 34.268, 30.57, 21.882, 19.071]
memory = [0.057, 559.18, 283.316, 307.94, 319.847, 267.062]
percent_nodes = [0.53, 100.00, 62.96, 46.41, 26.32, 21.30]

# Gráfico de Nodos
plt.figure(figsize=(6, 4))
plt.bar(algorithms, nodes, color='skyblue')
plt.xlabel("Algorithm")
plt.ylabel("Nodes")
plt.title("Number of Nodes Explored by Algorithm")
plt.xticks(rotation=45)
plt.savefig("nodes.png", format="png", dpi=300)
plt.show()

# Gráfico de Tiempo
plt.figure(figsize=(6, 4))
plt.bar(algorithms, time, color='salmon')
plt.xlabel("Algorithm")
plt.ylabel("Time (s)")
plt.title("Execution Time by Algorithm")
plt.xticks(rotation=45)
plt.savefig("time.png", format="png", dpi=300)
plt.show()

# Gráfico de Memoria
plt.figure(figsize=(6, 4))
plt.bar(algorithms, memory, color='lightgreen')
plt.xlabel("Algorithm")
plt.ylabel("Memory (MB)")
plt.title("Memory Usage by Algorithm")
plt.xticks(rotation=45)
plt.savefig("memory.png", format="png", dpi=300)
plt.show()

# Gráfico de Porcentaje de Nodos
plt.figure(figsize=(6, 4))
plt.bar(algorithms, percent_nodes, color='orange')
plt.xlabel("Algorithm")
plt.ylabel("Percentage of Nodes (%)")
plt.title("Percentage of Nodes Explored by Algorithm")
plt.xticks(rotation=45)
plt.savefig("perc.png", format="png", dpi=300)

plt.show()
```
